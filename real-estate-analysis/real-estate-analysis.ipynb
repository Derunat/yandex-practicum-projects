## Предобработка данных

```python
#переименование столбца
df = df.rename(columns={'cityCenters_nearest':'city_centers_nearest'})
```

Меняем тип данных столбцов first_day_exposition и is_appartment. Остальные, в которых вместо ожидаемого int оказывается float менять не будем, потому что такой тип выходит из-за значений NaN в столбце

```python

df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'],format='%Y-%m-%dT%H:%M:%S')

df['is_apartment'] = df['is_apartment'].astype('boolean') #заменяем на boolean вместо bool, чтобы все NaN не прерватились в True
   
df.info()
df.head()
```

Пропуски есть в 11 столбцах, но заполнять уместно не все. Оставим без изменений пропуски в следующих столбцах:
 - ceiling_height
 - kitchen_area
 - airports_nearest
 - cityCenters_nearest
 - parks_around3000
 - ponds_around3000 
 - ponds_nearest

Этой информации человек может не знать (и не хотеть измерять) и просто не указать в объявлении. 

Пропуски в days_exposition вероятно означают, что объявление еще не снято. Возможно, в реальной задаче можно было бы заменить пропуски на число дней до текущей даты. В данной задаче оставим их без изменений.



В оставшихся колонках пропуски следует обработать

 - в is_apartment пропусков подавляющее большинство. Нужно посмотреть подробнее и понять что это значит
 - в balcony NaN заменим на 0
 - строки с пропуском в locality_name следует выкинуть. Название населенного пункта - критически важные данные для продажи жилья, поэтому заменить пропуски ничем нельзя. Их всего 49, это меньше 1%, таким объемом данных можно пожертвовать. 

 


```python
# Заполнение пропусков в столбце balcony на нули
df['balcony'] = df['balcony'].fillna(0) 
df['balcony'].isna().sum() # проверка
```

```python
#Удаление строк с незаполненным locality_name
df = df.dropna(subset=['locality_name'])
df.info() #проверка
```

Посмотрим как выглядят усредененные значния других колонок для объектов с True и False в столбце is_apartment

```python
df.pivot_table(columns='is_apartment')
```

Видно, что "open_plan" у квартир всегда равен False. То же относится к столбцу "studio". Других явных закономерностей обнаружить не получается. Следовало бы запросить больше данных, но в данной задаче можно заменить объекты с True в "open_plan" и "studio" на "не квартиры", а остальные пропуски в is_apartment заменить на True

```python
df.loc[df['studio']==True & df['is_apartment'].isna(),'is_apartment'] = False
df.loc[df['open_plan']==True & df['is_apartment'].isna(),'is_apartment'] = False
df['is_apartment']= df['is_apartment'].fillna(True)
```

В колонке с названиями населенных пунктов могут быть разные формы записи одних и тех же пунктов

```python
display(df['locality_name'].nunique())
df['locality_name'].sort_values().unique()

```

Дубликаты точно есть, как минимум "посёлки" и "поселки". Еще больше дубликатов найдется если совсем убрать типы населенных пунктов из названий, но тогда можем потерять пункты с одинаковым названием, но разного типа. В реальной задаче следовало бы и уточнить информацию. Сейчас просто удалим все уточняющие формулировки, так как одна из целей - определить топ-10 населенных пунктов. Чтобы сохранить информацию, сделаем это в новой колонке


```python
df['locality_name']=df['locality_name'].str.replace('посёлок', 'поселок') #переименовываем посёлки в поселки
df['locality_name_droped'] = df['locality_name'] #дублируем названия населенных пунктов
df['locality_name_droped']=df['locality_name_droped'].str.replace('село ', '')
df['locality_name_droped']=df['locality_name_droped'].str.replace('садовое товарищество ', '')
df['locality_name_droped']=df['locality_name_droped'].str.replace('садовое товарищество ', '')
df['locality_name_droped']=df['locality_name_droped'].str.replace('садоводческое некоммерческое товарищество ', '')
df['locality_name_droped']=df['locality_name_droped'].str.replace('поселок станции ', '')
df['locality_name_droped']=df['locality_name_droped'].str.replace('поселок при железнодорожной станции ', '')
df['locality_name_droped']=df['locality_name_droped'].str.replace('поселок городского типа ', '')
df['locality_name_droped']=df['locality_name_droped'].str.replace('коттеджный поселок ', '')
df['locality_name_droped']=df['locality_name_droped'].str.replace('городской поселок ', '')
df['locality_name_droped']=df['locality_name_droped'].str.replace('деревня ', '')
df['locality_name_droped']=df['locality_name_droped'].str.replace('поселок ', '')


 #проверяем сколько дубликатов удалось удалить
display(df['locality_name_droped'].nunique())
df['locality_name_droped'].sort_values().unique()
```

## Расчёты и добавление результатов в таблицу

```python
df['price_per_square'] = df['last_price']/df['total_area'] #цена квадратного метра
df['weekday'] = pd.DatetimeIndex(df['first_day_exposition']).weekday #день недели публикации объявления
df['month'] = pd.DatetimeIndex(df['first_day_exposition']).month #месяц недели публикации объявления
df['year'] = pd.DatetimeIndex(df['first_day_exposition']).year #год недели публикации объявления
df['living_per_total'] = df['living_area']/df['total_area'] #соотношение жилой и общей площади
df['kitchen_per_total'] = df['kitchen_area']/df['total_area'] #отношение площади кухни к общей

#функция для присвоения категории этажу
def floor_group(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
    
df['floor_group'] = df.apply(floor_group, axis =1)
```
